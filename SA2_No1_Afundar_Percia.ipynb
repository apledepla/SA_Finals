# Number 8

```python
months_replace={
    'Janeiro':'January',
    'Fevereiro':'February',
    'Março':'March',
    'Abril':'April',
    'Maio': 'May',
    'Junho':'June',
    'Julho':'July',
    'Agosto':'August',
    'Setembro':'September',
    'Outubro':'October',
    'Novembro':'November',
    'Dezembro':'December'
    
}

dp.loc[:, 'month'] = dp['month'].replace(months_replace)

dp
```

# Number 9

```python
sum_of_fires=dp['number'].sum()    

print(sum_of_fires)
```

# Number 10

```python
monthly_sums = dp.groupby('month')['number'].sum()

largest_month = monthly_sums.idxmax()
largest_sum = monthly_sums.max()

print(f"{largest_month} has the largest fire incidents with: {largest_sum}")
```

# Number 11

```python
yearly_sums = dp.groupby('year')['number'].sum()

largest_year=yearly_sums.idxmax()
largest_sum_year = yearly_sums.max()

print(f"{largest_year} has the largest fire incidents with: {largest_sum_year}")
```

# Number 12

```python
state_sums = dp.groupby('state')['number'].sum()

largest_state=state_sums.idxmax()
largest_sum_state = state_sums.max()

print(f"{largest_state} has the largest fire incidents with: {largest_sum_state}")
```
